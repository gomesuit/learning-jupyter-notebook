```python
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)

y = x + np.random.randn(100)

plt.plot(x, y)
plt.show()
```

```python
# 散布図

plt.scatter(x, y)
plt.show()
```

```python
# ヒストグラム
plt.hist(np.random.randn(1000), bins=10)
plt.show()
```

```python
# 色

target = np.array([0 for i in range(50)] + [1 for i in range(50)])
```

```python
plt.scatter(x, y, c=target)
plt.show()
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
plt.hist(dataset2)
```

```python
plt.hist(dataset1, density=True, alpha=0.5, bins=20)
plt.hist(dataset2, density=True, alpha=0.5, bins=20, color='indianred')
```
